# PCA for particles

```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,50)
y = 0.5*x+10+np.random.normal(loc=0, scale=10, size=50)
pts = np.array([x,y]).T
# pts = np.random.normal(loc=0,scale=(5,4),size=(50,2))
```

```python
plt.scatter(pts[:,0],pts[:,1])
```

```python
pca= PCA(1)
pts_transformed = pca.fit_transform(pts)
pts_n = pca.inverse_transform(pts_transformed)
```
